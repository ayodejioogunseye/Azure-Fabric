**Reading Bronze Lakehouse data** 

```python
df1 = spark.read.format("csv")\
                .option("header",True)\
                .option("inferSchema", True)\
                .load("abfss://fabrictut@onelake.dfs.fabric.microsoft.com/DatacruxLakeHse.Lakehouse/Files/RawDataAdventureWorks_Customers")
```

```python
display(df1)
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

df1 = df1.withColumn("fullname",concat(col("prefix"),lit( " "),col("FirstName"),lit(" "),col("LastName")))
display(df1)
```

```python
df1 = df1.withColumn("Domain",split("EmailAddress","@")[1])
display(df1)
```

```python
df1.write.format("delta")\
 .mode("append")\
 .option("path","Tables/customers_3")\
 .saveAsTable("customers_3")
```
